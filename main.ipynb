```python
import os

def rename_files_in_folder(folder_path):
    # à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸§à¹ˆà¸²à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œà¸¡à¸µà¸­à¸¢à¸¹à¹ˆà¸ˆà¸£à¸´à¸‡à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
    if not os.path.exists(folder_path):
        print(f"Folder '{folder_path}' does not exist.")
        return

    # à¸”à¸¶à¸‡à¸£à¸²à¸¢à¸à¸²à¸£à¹„à¸Ÿà¸¥à¹Œà¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”à¹ƒà¸™à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œ
    files = os.listdir(folder_path)
    
    # à¸¥à¸¹à¸›à¹€à¸žà¸·à¹ˆà¸­à¹€à¸›à¸¥à¸µà¹ˆà¸¢à¸™à¸Šà¸·à¹ˆà¸­à¹„à¸Ÿà¸¥à¹Œ
    for index, filename in enumerate(files):
        # à¸ªà¸£à¹‰à¸²à¸‡à¹€à¸ªà¹‰à¸™à¸—à¸²à¸‡à¹„à¸Ÿà¸¥à¹Œà¹€à¸”à¸´à¸¡à¹à¸¥à¸°à¹„à¸Ÿà¸¥à¹Œà¹ƒà¸«à¸¡à¹ˆ
        old_file = os.path.join(folder_path, filename)
        file_extension = os.path.splitext(filename)[1]
        new_file = os.path.join(folder_path, f"{index}{file_extension}")

        # à¹€à¸›à¸¥à¸µà¹ˆà¸¢à¸™à¸Šà¸·à¹ˆà¸­à¹„à¸Ÿà¸¥à¹Œ
        os.rename(old_file, new_file)
        print(f"Renamed: {filename} -> {index}{file_extension}")

# à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸‡à¸²à¸™à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸žà¸£à¹‰à¸­à¸¡à¸£à¸°à¸šà¸¸à¹€à¸ªà¹‰à¸™à¸—à¸²à¸‡à¹‚à¸Ÿà¸¥à¹€à¸”à¸­à¸£à¹Œ
folder_path = "sample_images"
rename_files_in_folder(folder_path)
```
